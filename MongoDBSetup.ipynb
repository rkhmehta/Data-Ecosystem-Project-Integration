```python
#Importing libraries
import pandas as pd
import json
import pandas as pd
```

```python
#Connecting the data to the flipside API to extract data on NFTs

from flipside import Flipside

# Initialize `Flipside` with your API Key and API Url
flipside = Flipside("60fd2672-d86c-4e36-8d75-f3df00982394", "https://api-v2.flipsidecrypto.xyz")
```

```python
#Extracting Data on Sales From Flipside

sql_sales = """
WITH FirstTimeSaleTime AS (
  SELECT
    TOKENID,
    MIN(BLOCK_TIMESTAMP) AS FirstSaleTime
  FROM
    ethereum.nft.ez_nft_sales e
  WHERE
    e.PROJECT_NAME LIKE '%Pudgy Penguin%'
    OR e.PROJECT_NAME LIKE '%pudgy penguin%'
  GROUP BY
    TOKENID
)

SELECT
  e.AGGREGATOR_NAME,
  e.BLOCK_NUMBER,
  e.BLOCK_TIMESTAMP,
  e.BUYER_ADDRESS,
  e.CREATOR_FEE,
  e.CREATOR_FEE_USD,
  e.CURRENCY_ADDRESS,
  e.CURRENCY_SYMBOL,
  e.ERC1155_VALUE,
  e.EVENT_TYPE,
  e.NFT_ADDRESS,
  e.ORIGIN_FROM_ADDRESS,
  e.ORIGIN_FUNCTION_SIGNATURE,
  e.ORIGIN_TO_ADDRESS,
  e.PLATFORM_ADDRESS,
  e.PLATFORM_EXCHANGE_VERSION,
  e.PLATFORM_FEE,
  e.PLATFORM_FEE_USD,
  e.PLATFORM_NAME,
  e.PRICE,
  e.PRICE_USD,
  e.PROJECT_NAME,
  e.SELLER_ADDRESS,
  e.TOKEN_METADATA,
  e.TOKENID,
  e.TOTAL_FEES,
  e.TOTAL_FEES_USD,
  e.TX_FEE,
  e.TX_FEE_USD
FROM
  ethereum.nft.ez_nft_sales e
INNER JOIN  FirstTimeSaleTime f ON f.TOKENID = e.tokenid
WHERE
  (
    e.PROJECT_NAME LIKE '%Pudgy Penguin%'
    OR e.PROJECT_NAME LIKE '%pudgy penguin%'
  )
  AND e.EVENT_TYPE = 'sale'
  AND e.CURRENCY_SYMBOL = 'ETH'
  AND e.block_timestamp > f.FirstSaleTime
"""

# Run the query against Flipside's query engine and await the results
output_sales_query= flipside.query(sql_sales)
print(output_sales_query)
```

```python
# Converting the data from flipside from QueryResultSet to dataframe for sales database

df_c = output_sales_query.columns
df_col = pd.DataFrame([col[0] if isinstance(col, tuple) else col for col in df_c])

df_r = output_sales_query.rows
df_rows = pd.DataFrame(df_r)

df_rows.columns = df_col[0]

df_sales = df_rows

# Print the DataFrame
print(df_sales.head())
```
